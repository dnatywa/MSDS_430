Step 1: Using the random submodule in NumPy, create an array of 75 randomly generated integers between 1 and 15. Print the first 20 values.

```python
import numpy as np

random_int = np.random.randint(1,16,75)
print(random_int[0:20])
```

Step 2: Using the mean( ) and std( ) functions in NumPy, calculate the mean and standard deviation of your array.

```python
random_int_mean = random_int.mean()
print("mean:", str(random_int_mean))

random_int_std = np.std(random_int)
print("standard deviation:", str(random_int_std))
```

Step 3: Create an array of 75 randomly generated values from the normal distribution based on your mean and standard deviation. Print the first 20 values.

```python
normal_distribution = np.random.normal(random_int_mean, random_int_std, 75)
print(normal_distribution[0:20])
```

Step 4: Create a basic scatterplot of the array using matplotlib.pyplot.

```python
import matplotlib.pyplot as plt

x = np.arange(1,76,1)
plt.scatter(x, normal_distribution)
plt.show()
```

Step 5: Label the x-axis "Customer #', the y-axis 'Time in Minutes', and title the scatterplot 'Simulation of Wait Time'.

```python
plt.xlabel("Customer #")
plt.ylabel("Time in Minutes")
plt.title("Simulation of Wait Time")

plt.scatter(x, normal_distribution)
plt.show()
```
